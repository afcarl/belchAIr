## Change objects, that determine how state will change

```python
class Change(object):
    def __init__(self):
        pass
    
class GameStateChange(Change):
    def __init__(self):
        pass
        
class DiscardHand(GameStateChange):
    def __init__(self):
        pass
    
class Shuffle(GameStateChange):
    def __init__(self):
        pass
    
class TaigaBottom(GameStateChange):
    def __init__(self):
        pass
    
class AddGoblins(GameStateChange):
    def __init__(self, amount):
        self.amount = amount
        
class ChangeMana(GameStateChange):
    def __init__(self, c_dict):
        self.c_dict = c_dict

class DrawCard(GameStateChange):
    def __init__(self, number_cards=1):
        self.number_cards = number_cards
    
class CardStateChange(Change):
    def __init__(self, card):
        self.card = card
    
class ChangeZone(CardStateChange):
    def __init__(self, card, from_zone, to_zone):
        self.card = card
        self.from_zone = from_zone
        self.to_zone = to_zone

class Tapped(CardStateChange):
    def __init__(self, card):
        self.card = card
```

## Actions are the types of things the AI can choose from in a certain state

```python
class Action(object):
    def __init__(self):
        pass
    
class Tap(Action):
    def __init__(self):
        pass
    
class TapForMana(Action):
    def __init__(self, c_dict):
        self.c_dict = c_dict
    
class Play(Action):
    def __init__(self):
        pass
    
class PlayWith(Action):
    def __init__(self, c_dict):
        self.c_dict = c_dict
    
class Exile(Action):
    def __init__(self):
        pass
    
class PassPriority(Action):
    def __init__(self):
        pass
    
class PassTurn(Action):
    def __init__(self):
        pass
    
class ManamorphosePlay(Action):
    def __init__(self, c_dict):
        self.c_dict = c_dict
        
class Sacrifice(Action):
    def __init__(self):
        pass
    
class SacrificeForMana(Action):
    def __init__(self, c_dict):
        self.c_dict = c_dict
        
class CharbelchActivation(Action):
    def __init__(self):
        pass
    
#class PlayImprint(Action):
#    def __init__(self, card):
#        self.card = card
```

## Cards

```python
class Card(object):
    def __init__(self):
        self.tappable = False
    
class Taiga(Card):
    
    def __init__(self):
        self.name = 'Taiga'
        
    def actions(self, zone, number_untapped):
        possible_actions = []
        if zone == 'Hand':
            possible_actions.append(
                (
                    Play(),
                    (
                        ChangeZone(self.name, 'Hand', 'Battlefield')
                    )
                ))
        if zone == 'Battlefield':
            if number_untapped > 0:
                possible_actions.append(
                    (
                        TapForMana({'Red': 1}), 
                        (
                            Tapped(self.name),
                            ChangeMana({'Red': 1})
                        )
                    ))
                possible_actions.append(
                    (
                        TapForMana({'Green': 1}),
                        (
                            Tapped(self.name),
                            ChangeMana({'Green': 1})
                        )
                    ))
        return possible_actions

    def all_actions(self):
        return [(Play(), (ChangeZone(self.name, 'Hand', 'Battlefield'))),
                (TapForMana({'Red': 1}), (Tapped(self.name), ChangeMana({'Red': 1}))),
                (TapForMana({'Green': 1}), (Tapped(self.name), ChangeMana({'Green': 1})))]
    
class ElvishSpiritGuide(Card):
    
    def __init__(self):
        self.name = 'Elvish Spirit Guide'
        
    def actions(self, zone):
        possible_actions = []
        if zone == 'Hand':
            possible_actions.append(
                (
                    Exile(),
                    (
                        ChangeMana({'Green': 1}),
                        ChangeZone(self.name, 'Hand', 'Exile')
                    )
                ))
        return possible_actions
    
    def all_actions(self):
        return [(Exile(), (ChangeMana({'Green': 1}), ChangeZone(self.name, 'Hand', 'Exile')))]
    
class SimianSpiritGuide(Card):
    
    def __init__(self):
        self.name = 'Simian Spirit Guide'
        
    def actions(self, zone):
        possible_actions = []
        if zone == 'Hand':
            possible_actions.append(
                (
                    Exile(),
                    (
                        ChangeMana({'Red': 1}),
                        ChangeZone(self.name, 'Hand', 'Exile')
                    )
                ))
        return possible_actions

    def all_actions(self):
        return [(Exile(), (ChangeMana({'Red': 1}), ChangeZone(self.name, 'Hand', 'Exile')))]
    
class TinderWall(Card):

    def __init__(self):
        self.name = 'Tinder Wall'
        
    def actions(self, zone, mana_pool):
        possible_actions = []
        if zone == 'Hand':
            if mana_pool['Green'] > 0:
                possible_actions.append(
                    (
                        Play(), 
                        (
                            ChangeZone(self.name, 'Hand', 'Battlefield'),
                            ChangeMana({'Green': -1})
                        )
                    ))
        if zone == 'Battlefield':
            possible_actions.append(
                (
                    Sacrifice(),
                    (
                        ChangeMana({'Red': 2}),
                        ChangeZone(self.name, 'Battlefield', 'Graveyard')
                    )
                ))
        return possible_actions
    
    def all_actions(self):
        return [(Play(), (ChangeZone(self.name, 'Hand', 'Battlefield'), ChangeMana({'Green': -1}))),
                (Sacrifice(), (ChangeMana({'Red': 2}), ChangeZone(self.name, 'Battlefield', 'Graveyard')))]
    
class DesperateRitual(Card):

    def __init__(self):
        self.name = 'Desperate Ritual'
        
    def actions(self, zone, mana_pool):
        possible_actions = []
        if zone == 'Hand':
            if mana_pool['Red'] > 1:
                possible_actions.append(
                    (
                        PlayWith({'Red': 2}),
                        (
                            ChangeMana({'Red': 1}),
                            ChangeZone(self.name, 'Hand', 'Graveyard')
                        )
                    ))
            if mana_pool['Red'] > 0 and mana_pool['Green'] > 0:
                possible_actions.append(
                    (
                        PlayWith({'Red': 1, 'Green': 1}),
                        (
                            ChangeMana({'Red': 2, 'Green': -1}),
                            ChangeZone(self.name, 'Hand', 'Graveyard')
                        )
                    ))
        return possible_actions
    
    def all_actions(self):
        return [(PlayWith({'Red': 2}), (ChangeMana({'Red': 1}), ChangeZone(self.name, 'Hand', 'Graveyard'))),
                (PlayWith({'Red': 1, 'Green': 1}), (ChangeMana({'Red': 2, 'Green': -1}), ChangeZone(self.name, 'Hand', 'Graveyard')))]
    
class PyreticRitual(Card):

    def __init__(self):
        self.name = 'Pyretic Ritual'
        
    def actions(self, zone, mana_pool):
        possible_actions = []
        if zone == 'Hand':
            if mana_pool['Red'] > 1:
                possible_actions.append(
                    (
                        PlayWith({'Red': 2}),
                        (
                            ChangeMana({'Red': 1}),
                            ChangeZone(self.name, 'Hand', 'Graveyard')
                        )
                    ))
            if mana_pool['Red'] > 0 and mana_pool['Green'] > 0:
                possible_actions.append(
                    (
                        PlayWith({'Red': 1, 'Green': 1}),
                        (
                            ChangeMana({'Red': 2, 'Green': -1}),
                            ChangeZone(self.name, 'Hand', 'Graveyard')
                        )
                    ))
        return possible_actions

    def all_actions(self):
        return [(PlayWith({'Red': 2}), (ChangeMana({'Red': 1}), ChangeZone(self.name, 'Hand', 'Graveyard'))),
                (PlayWith({'Red': 1, 'Green': 1}), (ChangeMana({'Red': 2, 'Green': -1}), ChangeZone(self.name, 'Hand', 'Graveyard')))]
    
class RiteOfFlame(Card):

    def __init__(self):
        self.name = 'Rite of Flame'
        
    def actions(self, zone, number_rites, mana_pool):
        possible_actions = []
        if zone == 'Hand':
            if mana_pool['Red'] > 0:
                possible_actions.append(
                    (
                        PlayWith({'Red': 1}),
                        (
                            ChangeMana({'Red': 1 + number_rites}),
                            ChangeZone(self.name, 'Hand', 'Graveyard')
                        )
                    ))
        return possible_actions
    
    def all_actions(self):
        return [(PlayWith({'Red': 1}), (ChangeMana({'Red': 1}), ChangeZone(self.name, 'Hand', 'Graveyard')))]
    
class SeethingSong(Card):

    def __init__(self):
        self.name = 'Seething Song'
        
    def actions(self, zone, mana_pool):
        possible_actions = []
        if zone == 'Hand':
            if mana_pool['Red'] > 2:
                possible_actions.append(
                    (
                        PlayWith({'Red': 3}),
                        (
                            ChangeMana({'Red': 2}),
                            ChangeZone(self.name, 'Hand', 'Graveyard')
                        )
                    ))
            if mana_pool['Red'] > 1 and mana_pool['Green'] > 0:
                possible_actions.append(
                    (
                        PlayWith({'Red': 2, 'Green': 1}),
                        (
                            ChangeMana({'Red': 3, 'Green': -1}),
                            ChangeZone(self.name, 'Hand', 'Graveyard')
                        )
                    ))
            if mana_pool['Red'] > 0 and mana_pool['Green'] > 1:
                possible_actions.append(
                    (
                        PlayWith({'Red': 1, 'Green': 2}),
                        (
                            ChangeMana({'Red': 4, 'Green': -2}),
                            ChangeZone(self.name, 'Hand', 'Graveyard')
                        )
                    ))
        return possible_actions
        

    def all_actions(self):
        return [(PlayWith({'Red': 3}), (ChangeMana({'Red': 2}), ChangeZone(self.name, 'Hand', 'Graveyard'))),
                (PlayWith({'Red': 2, 'Green': 1}), (ChangeMana({'Red': 3, 'Green': -1}), ChangeZone(self.name, 'Hand', 'Graveyard'))),
                (PlayWith({'Red': 1, 'Green': 2}), (ChangeMana({'Red': 4, 'Green': -2}), ChangeZone(self.name, 'Hand', 'Graveyard')))]
    
class EmptyTheWarrens(Card):

    def __init__(self):
        self.name = 'Empty the Warrens'
        
    def actions(self, zone, storm_count):
        possible_actions = []
        if zone == 'Hand':
            if mana_pool['Red'] > 3:
                possible_actions.append(
                    (
                        PlayWith({'Red': 4}),
                        (
                            ChangeMana({'Red': -4}),
                            ChangeZone(self.name, 'Hand', 'Graveyard'),
                            AddGoblins(storm_count * 2)
                        )
                    ))
            if mana_pool['Red'] > 2 and mana_pool['Green'] > 0:
                possible_actions.append(
                    (
                        PlayWith({'Red': 3, 'Green': 1}),
                        (
                            ChangeMana({'Red': -3, 'Green': -1}),
                            ChangeZone(self.name, 'Hand', 'Graveyard'),
                            AddGoblins(storm_count * 2))
                    ))
            if mana_pool['Red'] > 1 and mana_pool['Green'] > 1:
                possible_actions.append(
                    (
                        PlayWith({'Red': 2, 'Green': 2}),
                        (
                            ChangeMana({'Red': -2, 'Green': -2}),
                            ChangeZone(self.name, 'Hand', 'Graveyard'),
                            AddGoblins(storm_count * 2)
                        )
                    ))
            if mana_pool['Red'] > 0 and mana_pool['Green'] > 2:
                possible_actions.append(
                    (
                        PlayWith({'Red': 1, 'Green': 3}),
                        (
                            ChangeMana({'Red': -1, 'Green': -3}),
                            ChangeZone(self.name, 'Hand', 'Graveyard'),
                            AddGoblins(storm_count * 2)
                        )
                    ))
        return possible_actions
    
    def all_actions(self):
        return [(PlayWith({'Red': 4}), (ChangeMana({'Red': -4}), ChangeZone(self.name, 'Hand', 'Graveyard'), AddGoblins(2))), 
                (PlayWith({'Red': 3, 'Green': 1}), (ChangeMana({'Red': -3, 'Green': -1}), ChangeZone(self.name, 'Hand', 'Graveyard'), AddGoblins(2))),
                (PlayWith({'Red': 2, 'Green': 2}), (ChangeMana({'Red': -2, 'Green': -2}), ChangeZone(self.name, 'Hand', 'Graveyard'), AddGoblins(2))),
                (PlayWith({'Red': 1, 'Green': 3}), (ChangeMana({'Red': -1, 'Green': -3}), ChangeZone(self.name, 'Hand', 'Graveyard'), AddGoblins(2)))]
    
class GitaxianProbe(Card):

    def __init__(self):
        self.name = 'Gitaxian Probe'
        
    def action(self):
        possible_actions = []
        if zone == 'Hand':
            possible_actions.append(
                (
                    Play(),
                    (
                        DrawCard(),
                        ChangeZone(self.name, 'Hand', 'Graveyard')
                    )
                ))
        return possible_actions
    
    def all_actions(self):
        return [(Play(), (DrawCard(), ChangeZone(self.name, 'Hand', 'Graveyard')))]
    
class LandGrant(Card):
    
    def __init__(self):
        self.name = 'Land Grant'
        
    def action(self, zone, number_taigas_hand, number_taigas_deck):
        possible_actions = []
        if zone == 'Hand':
            if number_taigas_deck > 0:
                possible_actions.append(
                    (
                        Play(),
                        (
                            ChangeZone(self.name, 'Hand', 'Graveyard'),
                            ChangeZone('Taiga', 'Deck', 'Hand'), Shuffle()
                        )
                    ))
            if number_taigas_deck == 0 and number_taigas_hand == 0:
                possible_actions.append(
                    (
                        Play(),
                        (
                            ChangeZone(self.name, 'Hand', 'Graveyard'),
                            Shuffle()
                        )
                    ))
        return possible_actions
    
    def all_actions(self):
        return [(Play(), (ChangeZone(self.name, 'Hand', 'Graveyard'), ChangeZone('Taiga', 'Deck', 'Hand'), Shuffle())),
                (Play(), (ChangeZone(self.name, 'Hand', 'Graveyard'), Shuffle()))]

class LotusPetal(Card):

    def __init__(self):
        self.name = 'Lotus Petal'
        
    def action(self, zone):
        possible_actions = []
        if zone == 'Hand':
            possible_actions.append(
                (
                    Play(),
                    (
                        ChangeZone(self.name, 'Hand', 'Battlefield'),
                    )
                ))
        if zone == 'Battlefield':
            possible_actions.append(
                (
                    SacrificeForMana({'Red': 1}),
                    (
                        ChangeZone(self.name, 'Battlefield', 'Graveyard'),
                        ChangeMana({'Red': 1})
                    )
                ))
            possible_actions.append(
                (
                    SacrificeForMana({'Green': 1}),
                    (
                        ChangeZone(self.name, 'Battlefield', 'Graveyard'),
                        ChangeMana({'Green': 1})
                    )
                ))
        return possible_actions
        
    def all_actions(self):
        return [(Play(), (ChangeZone(self.name, 'Hand', 'Battlefield'), )),
                (SacrificeForMana({'Red': 1}), (ChangeZone(self.name, 'Battlefield', 'Graveyard'), ChangeMana({'Red': 1}))),
                (SacrificeForMana({'Green': 1}), (ChangeZone(self.name, 'Battlefield', 'Graveyard'), ChangeMana({'Green': 1})))]

class LionsEyeDiamond(Card):

    def __init__(self):
        self.name = 'Lions Eye Diamond'
        
    def action(self, zone):
        possible_actions = []
        if zone == 'Hand':
            possible_actions.append(
                (
                    Play(),
                    (
                        ChangeZone(self.name, 'Hand', 'Battlefield'),
                    )
                ))
        if zone == 'Battlefield':
            possible_actions.append(
                (
                    SacrificeForMana({'Red': 3}),
                        (
                            ChangeZone(self.name, 'Battlefield', 'Graveyard'),
                            ChangeMana({'Red': 3}), DiscardHand()
                        )
                ))
            possible_actions.append(
                (
                    SacrificeForMana({'Green': 3}),
                    (
                        ChangeZone(self.name, 'Battlefield', 'Graveyard'),
                        ChangeMana({'Green': 3}),
                        DiscardHand()
                    )
                ))
        return possible_actions
    
    def all_actions(self):
        return [(Play(), (ChangeZone(self.name, 'Hand', 'Battlefield'), )), 
                (SacrificeForMana({'Red': 3}), (ChangeZone(self.name, 'Battlefield', 'Graveyard'), ChangeMana({'Red': 3}), DiscardHand())),
                (SacrificeForMana({'Green': 3}), (ChangeZone(self.name, 'Battlefield', 'Graveyard'), ChangeMana({'Green': 3}), DiscardHand()))]
    
class Manamorphose(Card):
    def __init__(self):
        self.name = 'Manamorphose'
        
    def action(self, zone, mana_pool):
        possible_actions = []
        if zone == 'Hand':
            if mana_pool['Red'] > 1:
                possible_actions.append(
                    (
                        ManamorphosePlay({'Red': 2}),
                        (
                            ChangeZone(self.name, 'Hand', 'Graveyard'),
                            DrawCard()
                        )
                    ))
                possible_actions.append(
                    (
                        ManamorphosePlay({'Red': 1, 'Green': 1}),
                        (
                            ChangeMana({'Red': -1, 'Green': 1}),
                            ChangeZone(self.name, 'Hand', 'Graveyard'),
                            DrawCard()
                        )
                    ))
                possible_actions.append(
                    (
                        ManamorphosePlay({'Green': 2}),
                        (
                            ChangeMana({'Red': -2, 'Green': 2}),
                            ChangeZone(self.name, 'Hand', 'Graveyard'),
                            DrawCard()
                        )
                    ))
            if mana_pool['Green'] > 1:
                possible_actions.append(
                    (
                        ManamorphosePlay({'Green': 2}),
                        (
                            ChangeZone(self.name, 'Hand', 'Graveyard'),
                            DrawCard()
                        )
                    ))
                possible_actions.append(
                    (
                        ManamorphosePlay({'Red': 1, 'Green': 1}),
                        (
                            ChangeMana({'Red': 1, 'Green': -1}),
                            ChangeZone(self.name, 'Hand', 'Graveyard'),
                            DrawCard()
                        )
                    ))
                possible_actions.append(
                    (
                        ManamorphosePlay({'Red': 2}), 
                        (
                            ChangeMana({'Red': 2, 'Green': -2}),
                            ChangeZone(self.name, 'Hand', 'Graveyard'),
                            DrawCard()
                        )
                    ))
            if mana_pool['Red'] > 0 and mana_pool['Green'] > 0:
                possible_actions.append(
                    (
                        ManamorphosePlay({'Red': 2}),
                        (
                            ChangeMana({'Red': 1, 'Green': -1}),
                            ChangeZone(self.name, 'Hand', 'Graveyard'),
                            DrawCard()
                        )
                    ))
                possible_actions.append(
                    (
                        ManamorphosePlay({'Red': 1, 'Green': 1}),
                        (
                            ChangeZone(self.name, 'Hand', 'Graveyard'),
                            DrawCard()
                        )
                    ))
                possible_actions.append(
                    (
                        ManamorphosePlay({'Green': 2}),
                        (
                            ChangeMana({'Red': -1, 'Green': 1}),
                            ChangeZone(self.name, 'Hand', 'Graveyard'),
                            DrawCard()
                        )
                    ))
        return possible_actions
    
    def all_actions(self):
        possible_actions = []
        possible_actions.append((ManamorphosePlay({'Red': 2}), (ChangeZone(self.name, 'Hand', 'Graveyard'), DrawCard())))
        possible_actions.append((ManamorphosePlay({'Red': 1, 'Green': 1}), (ChangeMana({'Red': -1, 'Green': 1}), ChangeZone(self.name, 'Hand', 'Graveyard'), DrawCard())))
        possible_actions.append((ManamorphosePlay({'Green': 2}), (ChangeMana({'Red': -2, 'Green': 2}), ChangeZone(self.name, 'Hand', 'Graveyard'), DrawCard())))
        possible_actions.append((ManamorphosePlay({'Green': 2}), (ChangeZone(self.name, 'Hand', 'Graveyard'), DrawCard())))
        possible_actions.append((ManamorphosePlay({'Red': 1, 'Green': 1}), (ChangeMana({'Red': 1, 'Green': -1}), ChangeZone(self.name, 'Hand', 'Graveyard'), DrawCard())))
        possible_actions.append((ManamorphosePlay({'Red': 2}), (ChangeMana({'Red': 2, 'Green': -2}), ChangeZone(self.name, 'Hand', 'Graveyard'), DrawCard())))
        possible_actions.append((ManamorphosePlay({'Red': 2}), (ChangeMana({'Red': 1, 'Green': -1}), ChangeZone(self.name, 'Hand', 'Graveyard'), DrawCard())))
        possible_actions.append((ManamorphosePlay({'Red': 1, 'Green': 1}), (ChangeZone(self.name, 'Hand', 'Graveyard'), DrawCard())))
        possible_actions.append((ManamorphosePlay({'Green': 2}), (ChangeMana({'Red': -1, 'Green': 1}), ChangeZone(self.name, 'Hand', 'Graveyard'), DrawCard())))
        return possible_actions

class GoblinCharbelcher(Card):
    def __init__(self):
        self.name = 'Goblin Charbelcher'
        
    def actions(self, zone, mana_pool):
        possible_actions = []
        if zone == 'Hand':
            if mana_pool['Red'] + mana_pool['Green'] > 3:
                if mana_pool['Red'] > 3:
                    possible_actions.append(
                        (
                            Play(),
                            (
                                ChangeZone(self.name, 'Hand', 'Battlefield'),
                                ChangeMana({'Red': -4})
                            )
                        ))
                if mana_pool['Red'] > 2 and mana_pool['Green'] > 0:
                    possible_actions.append(
                        (
                            Play(),
                            (
                                ChangeZone(self.name, 'Hand', 'Battlefield'),
                                ChangeMana({'Red': -3, 'Green': -1})
                            )
                        ))
                if mana_pool['Red'] > 1 and mana_pool['Green'] > 1:
                    possible_actions.append(
                        (
                            Play(),
                            (
                                ChangeZone(self.name, 'Hand', 'Battlefield'),
                                ChangeMana({'Red': -2, 'Green': -2})
                            )
                        ))
                if mana_pool['Red'] > 0 and mana_pool['Green'] > 2:
                    possible_actions.append(
                        (
                            Play(),
                            (
                                ChangeZone(self.name, 'Hand', 'Battlefield'),
                                ChangeMana({'Red': -1, 'Green': -3})
                            )
                        ))
                if mana_pool['Green'] > 3:
                    possible_actions.append(
                        (
                            Play(),
                            (
                                ChangeZone(self.name, 'Hand', 'Battlefield'),
                                ChangeMana({'Green': -4})
                            )
                        ))
        
        if zone == 'Battlefield':
            if mana_pool['Red'] + mana_pool['Green'] > 2:
                if mana_pool['Red'] > 2:
                    possible_actions.append(
                        (
                            CharbelchActivation(),
                            (
                                ChangeMana({'Red': -3}),
                                Tapped('GoblinCharbelcher')
                            )
                        ))
                if mana_pool['Red'] > 1 and mana_pool['Green'] > 0:
                    possible_actions.append(
                        (
                            CharbelchActivation(),
                            (
                                ChangeMana({'Red': -2, 'Green': -1}),
                                Tapped('GoblinCharbelcher')
                            )
                        ))
                if mana_pool['Red'] > 0 and mana_pool['Green'] > 1:
                    possible_actions.append(
                        (
                            CharbelchActivation(),
                            (
                                ChangeMana({'Red': -1, 'Green': -2}),
                                Tapped('GoblinCharbelcher')
                            )
                        ))
                if mana_pool['Green'] > 2:
                    possible_actions.append(
                        (
                            CharbelchActivation(),
                            (
                                ChangeMana({'Green': -3}),
                                Tapped('GoblinCharbelcher')
                            )
                        ))
        return possible_actions
    
    def all_actions(self):
        return [(Play(), (ChangeZone(self.name, 'Hand', 'Battlefield'), ChangeMana({'Red': -4}))),
                (Play(), (ChangeZone(self.name, 'Hand', 'Battlefield'), ChangeMana({'Red': -3, 'Green': -1}))),
                (Play(), (ChangeZone(self.name, 'Hand', 'Battlefield'), ChangeMana({'Red': -2, 'Green': -2}))),
                (Play(), (ChangeZone(self.name, 'Hand', 'Battlefield'), ChangeMana({'Red': -1, 'Green': -3}))),
                (Play(), (ChangeZone(self.name, 'Hand', 'Battlefield'), ChangeMana({'Green': -4}))),
                (CharbelchActivation(), (ChangeMana({'Red': -3}), Tapped('GoblinCharbelcher'))),
                (CharbelchActivation(), (ChangeMana({'Red': -2, 'Green': -1}), Tapped('GoblinCharbelcher'))),
                (CharbelchActivation(), (ChangeMana({'Red': -1, 'Green': -2}), Tapped('GoblinCharbelcher'))),
                (CharbelchActivation(), (ChangeMana({'Green': -3}), Tapped('GoblinCharbelcher')))]

class ChromeMox(Card):    
    def __init__(self):
        self.name = 'Chrome Mox'
        
    def actions(self, zone, hand, untapped_moxen):
        """
        List of all possible actions given the zone and the list of unique card names
        
        :param str zone: Zone the card exists in
        :param list hand: List of all non-mox card names that are currently in the hand
        :param int untapped_moxen: Number of untapped moxen
        :returns: List of (Action, Changes) tuple
        """
        possible_actions = []
        if zone == 'Hand':
            if len(hand) > 0:
                for card in card_names:
                    if card in hand:
                        possible_actions.append(
                            (
                                Play(),
                                (
                                    ChangeZone(self.name, 'Hand', 'Battlefield'),
                                    ChangeZone(card, 'Hand', 'Exile')
                                )
                            ))
        if zone == 'Battlefield':
            if untapped_moxen > 0:
                possible_actions.append(
                    (
                        Tap(),
                        (
                            ChangeMana({'Red': 1}),
                            Tapped('ChromeMox')
                        )
                    ))
                possible_actions.append(
                    (
                        Tap(),
                        (
                            ChangeMana({'Green': 1}),
                            Tapped('ChromeMox')
                        )
                    ))
        return possible_actions
    
    def all_actions(self):
        possible_actions = []
        for card in card_names:
            possible_actions.append((Play(), (ChangeZone(self.name, 'Hand', 'Battlefield'), ChangeZone(card, 'Hand', 'Exile'))))
        possible_actions.append((Tap(), (ChangeMana({'Red': 1}), Tapped('ChromeMox'))))
        possible_actions.append((Tap(), (ChangeMana({'Green': 1}), Tapped('ChromeMox'))))
        return possible_actions
    
class BurningWish(Card):
    def __init__(self):
        self.name = 'Burning Wish'
        
    def actions(self, zone, mana_pool):
        possible_actions = []
        if zone == 'Hand':
            if mana_pool['Red'] > 1:
                for wishable_card in wishable_cards:
                    possible_actions.append(
                        (
                            Play(),
                            (
                                ChangeMana({'Red': -2}),
                                ChangeZone(self.name, 'Hand', 'Exile'),
                                ChangeZone(wishable_card, 'Exile', 'Hand')
                            )
                        ))
            if mana_pool['Red'] > 0 and mana_pool['Green'] > 0:
                for wishable_card in wishable_cards:
                    possible_actions.append(
                        (
                            Play(),
                            (
                                ChangeMana({'Red': -1, 'Green': -1}),
                                ChangeZone(self.name, 'Hand', 'Exile'),
                                ChangeZone(wishable_card, 'Exile', 'Hand')
                            )
                        ))
        return possible_actions
    
    def all_actions(self):
        possible_actions = []
        for wishable_card in wishable_cards:
            possible_actions.append((Play(), (ChangeMana({'Red': -2}), ChangeZone(self.name, 'Hand', 'Exile'), ChangeZone(wishable_card, 'Exile', 'Hand'))))
        for wishable_card in wishable_cards:
            possible_actions.append((Play(), (ChangeMana({'Red': -1, 'Green': -1}), ChangeZone(self.name, 'Hand', 'Exile'), ChangeZone(wishable_card, 'Exile', 'Hand'))))
        return possible_actions
    
class ReforgeTheSoul(Card):
    def __init__(self):
        self.name = 'Reforge the Soul'
        
    def actions(self, zone):
        possible_actions = []
        if zone == 'Hand':
            if mana_pool['Red'] > 4:
                possible_actions.append(
                    (
                        Play(),
                        (
                            ChangeMana({'Red': -5}),
                            ChangeZone(self.name, 'Hand', 'Graveyard'),
                            DiscardHand(),
                            DrawCard(7)
                        )
                    ))
            if mana_pool['Red'] > 3 and mana_pool['Green'] > 0:
                possible_actions.append(
                    (
                        Play(),
                        (
                            ChangeMana({'Red': -4, 'Green': -1}),
                            ChangeZone(self.name, 'Hand', 'Graveyard'),
                            DiscardHand(), 
                            DrawCard(7)
                        )
                    ))
            if mana_pool['Red'] > 2 and mana_pool['Green'] > 1:
                possible_actions.append(
                    (
                        Play(),
                        (
                            ChangeMana({'Red': -3, 'Green': -2}),
                            ChangeZone(self.name, 'Hand', 'Graveyard'),
                            DiscardHand(), 
                            DrawCard(7)
                        )
                    ))
            if mana_pool['Red'] > 1 and mana_pool['Green'] > 2:
                possible_actions.append(
                    (
                        Play(),
                        (
                            ChangeMana({'Red': -2, 'Green': -3}),
                            ChangeZone(self.name, 'Hand', 'Graveyard'),
                            DiscardHand(), 
                            DrawCard(7)
                        )
                    ))
        return possible_actions
    
    def all_actions(self):
        possible_actions = []
        possible_actions.append((Play(), (ChangeMana({'Red': -5}), ChangeZone(self.name, 'Hand', 'Graveyard'), DiscardHand(), 
                                          DrawCard(7))))
        possible_actions.append((Play(), (ChangeMana({'Red': -4, 'Green': -1}), ChangeZone(self.name, 'Hand', 'Graveyard'), DiscardHand(), 
                                          DrawCard(7))))
        possible_actions.append((Play(), (ChangeMana({'Red': -3, 'Green': -2}), ChangeZone(self.name, 'Hand', 'Graveyard'), DiscardHand(), 
                                          DrawCard(7))))
        possible_actions.append((Play(), (ChangeMana({'Red': -2, 'Green': -3}), ChangeZone(self.name, 'Hand', 'Graveyard'), DiscardHand(), 
                                          DrawCard(7))))
        return possible_actions
```
